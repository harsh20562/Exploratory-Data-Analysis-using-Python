## Exploratory Analysis and Visualization

**TODO** - In this, I will compute the sum of crimes against crimes in various states and districts and will tell about the sum and mean of cases for some states


Let's begin by importing`matplotlib.pyplot` and `seaborn`.

````python
import seaborn as sns
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 9)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
````

````python
year_df=crime_df.groupby('Year')[['Rape','Kidnapping and Abduction','Dowry Deaths','Assault on women with intent to outrage her modesty','Insult to modesty of Women','Cruelty by Husband or his Relatives','Importation of Girls']].sum()
````

Here I am grouping my dataset on the factor Year that is Year will act as index from year_df and I used sum() to get total cases against women crime for a particular year

````python
year_df
````

````python
plt.xlabel('Year');
plt.ylabel('Cases Reported');
plt.plot(year_df.index,year_df['Rape'],marker='o');
plt.title('Rape');
````

In the above plot, we can see how rape cases increased in India from 2001 to 2014.
In this we plotted a line graph for seeing the increase in rape cases for different years.

````python
plt.xlabel('Year');
plt.ylabel('Cases Reported');
plt.plot(year_df.index,year_df['Dowry Deaths'],marker='o');
plt.title('Dowry Deaths');
````

India is a developing country and we are living in 21st century and still in India, we have so many cases of Dowry Deaths which is very shameful for such a country like India where women are worshipped as Goddess. This is all due to not having strict laws for women safety in India.

````python
plt.bar(year_df.index,year_df['Cruelty by Husband or his Relatives']);
plt.xlabel('Year');
plt.ylabel('Cases Registered');
plt.title('Cruelty by Husband or his Relatives');
````

````python
crime_df['STATE/UT'] = crime_df['STATE/UT'].str.replace('A&N ISLANDS','A & N ISLANDS')
crime_df['STATE/UT'] = crime_df['STATE/UT'].str.replace('DELHI UT','DELHI')
crime_df['STATE/UT'] = crime_df['STATE/UT'].str.replace('D&N HAVELI','D & N HAVELI')
crime_df = crime_df.replace('\t','', regex=True)
````

In this we are removing the \t from the back of strings
Also we are removing all occurrences of a particular value in a column by another value using replace() function

````python
crime_df
````

````python
state_df=crime_df.groupby('STATE/UT')['Rape','Kidnapping and Abduction','Dowry Deaths','Assault on women with intent to outrage her modesty','Insult to modesty of Women','Cruelty by Husband or his Relatives','Importation of Girls'].sum()
````

In this we are grouping it on the basis of STATE/UT so that we can analyse the crimes against women statewise.

````python
state_df.shape
````

It tells the dimension for the dataframe state_df

````python
state_df
````

````python
cm=['Blues','Reds','Greens','Oranges','Purples','Greys','twilight']
f, axs = plt.subplots(1, state_df.columns.size, gridspec_kw={'wspace': 0})
for i, (s, a, c) in enumerate(zip(state_df.columns, axs, cm)):
    sns.heatmap(np.array([state_df[s].values]).T, yticklabels=state_df.index, xticklabels=[s], annot=True, fmt='.2f', ax=a, cmap=c, cbar=False)
    if i>0:
        a.yaxis.set_ticks([])
        a.set_xticklabels(a.get_xticklabels(), rotation=90)
````

In this we used seaborn library of python. In the above cell we plot a heatmap which indicates the women crimes in different states and UT of India and we used different colours for different columns.
I also rotated the values of x axis by 90 degrees as they were overlapping.

````python
plt.plot(state_df.index,state_df.Rape);
plt.xticks(rotation=90);
plt.xlabel('States')
plt.ylabel('Cases Reported');
plt.title('Total Rape Cases In India from 2001-2014');
````

````python
fig,axes=plt.subplots(2,3,figsize=(20,20));
#For subplot 1
axes[0,0].plot(state_df.index,state_df['Kidnapping and Abduction']);
axes[0,0].set_title('Kidnapping and Abduction');
axes[0,0].set_xlabel('STATES');
axes[0,0].set_ylabel('Cases Reported');
axes[0,0].set_xticklabels(labels=state_df.index,Rotation=90);

#For subplot 2
axes[0,1].plot(state_df.index,state_df['Dowry Deaths']);
axes[0,1].set_title('Dowry Deaths');
axes[0,1].set_xlabel('STATES');
axes[0,1].set_ylabel('Cases Reported');
axes[0,1].set_xticklabels(labels=state_df.index,Rotation=90);

#For subplot 3
axes[0,2].plot(state_df.index,state_df['Assault on women with intent to outrage her modesty']);
axes[0,2].set_title('Assault on women with intent to outrage her modesty');
axes[0,2].set_xlabel('STATES');
axes[0,2].set_ylabel('Cases Reported');
axes[0,2].set_xticklabels(labels=state_df.index,Rotation=90);

#For subplot 4
axes[1,0].plot(state_df.index,state_df['Insult to modesty of Women']);
axes[1,0].set_title('Insult to modesty of Women');
axes[1,0].set_xlabel('STATES');
axes[1,0].set_ylabel('Cases Reported');
axes[1,0].set_xticklabels(labels=state_df.index,Rotation=90);

#For subplot 5
axes[1,1].plot(state_df.index,state_df['Cruelty by Husband or his Relatives']);
axes[1,1].set_title('Cruelty by Husband or his Relatives');
axes[1,1].set_xlabel('STATES');
axes[1,1].set_ylabel('Cases Reported');
axes[1,1].set_xticklabels(labels=state_df.index,Rotation=90);

#For ubplot 6
axes[1,2].plot(state_df.index,state_df['Importation of Girls']);
axes[1,2].set_title('Importation of Girls');
axes[1,2].set_xlabel('STATES');
axes[1,2].set_ylabel('Cases Reported');
axes[1,2].set_xticklabels(labels=state_df.index,Rotation=90);
plt.tight_layout(pad=2);
````

set_xlabel : It displays the parameter needed to be displayed on x-axis.
set_ylabel : It displays the parameter needed to be displayed on y-axis.
set_title : It sets the title for the plot 
tight_layout : It ensures the spacing between multiple subplots

#### Q1: TODO - Which state has the highest number of rape cases ?

#### Q2: TODO - Which year we had maximum number of Dowry Deaths cases ?

#### Q3: TODO - In which year there is maximum increase in number of cases of Dowry Death and how much ?

````python
diff_df=year_df.diff(axis=0)
````

diff(axis=0) gives us the difference between the values of adjacent rows. In this we are trying to analyse that in which year the cases for a particular evil increased the most so that we can analyse later why it happened 

````python
diff_df
````

````python
diff_df['Dowry Deaths'].max()
````

In this we are accessing the maximum value in a particular column

````python
diff_df.index[diff_df['Dowry Deaths']==1662.0]
````
